```python
# reversedns function
import socket
import requests

# IP address to perform reverse DNS lookup
# mit zdns ptr records finden
#ip_address = "79.124.49.86"

def reversedns(ip):
    try:
        # Perform reverse DNS lookup
        hostnames = socket.gethostbyaddr(ip)
        return hostnames    
    except socket.herror as e:
        #print("Reverse DNS lookup failed:", e)
        return None
```
